Task 1:
----
Load the data and replace missingd data with the mean value of all other data in this feature.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from autorank import autorank, create_report, plot_stats

%load_ext autoreload
%autoreload 2
RANDOM_STATE = 1234
np.random.seed(RANDOM_STATE)

import warnings
warnings.filterwarnings("ignore")
```

```python
X = pd.read_csv(os.path.join("data", "data_A2.csv"), header = None)
y = pd.read_csv(os.path.join("data", "labels_A2.csv"), header = None)
#apply SelectKBest class to extract top 10 best features
print("X is read")
print(X)
print("y read")
print(y)
```

Missing data
-----------

```python
# step 1: use SimpleImputer module to replace all the np.nan value with the mean value of the feature
imp = SimpleImputer()
imp.fit(X)
X_Fixed_Miss = imp.transform(X)
print("X after fixed missing data")
print(X_Fixed_Miss)
y = y.to_numpy()
```

Task 2:
----
Find the 10 most important features, using the Random Forest module.

```python

clf_RF = RandomForestClassifier(n_estimators=20, random_state=RANDOM_STATE, n_jobs=-1)
clf_RF.fit(X_Fixed_Miss, y)

feat_importances = pd.Series(clf_RF.feature_importances_, index=[i for i in range(100)])
indexs = feat_importances.nlargest(10).index
X_Top10_Features = X_Fixed_Miss[:, indexs]
print(X_Top10_Features.shape)
print(f"select features: {indexs}")
```

Task 3
----
Perform Random Forest, Decision tree, and pruned tree, and decision stamp, on the dataset from task2.

```python
# As for pruning, in order to find out the best hyper parameters for pruning, we can use cross validation to test the score for each parameter
# tuning max_depth
res = []
bestScore = 0
best_max_depth = 0
for i in range(1,20):
    # Kfold = 5
    kfold = StratifiedKFold(n_splits=5, random_state=RANDOM_STATE, shuffle=True)
    score = []
    for idx_train, idx_test in kfold.split(X_Top10_Features, y):
        X_train = X_Top10_Features[idx_train]
        X_test = X_Top10_Features[idx_test]
        y_train = y[idx_train]
        y_test = y[idx_test]
        model = DecisionTreeClassifier(max_depth=i, random_state = RANDOM_STATE)
        model.fit(X_train, y_train)
        score.append(model.score(X_test, y_test))
    mean = np.mean(score)
    res.append(mean)
    if mean > bestScore:
        best_max_depth = i
        bestScore = mean

plt.plot(range(1,20), res, color = "red", label = "depth")
plt.legend()
plt.show()

# tuning min_samples_leaf
res = []
bestScore = 0
best_min_samples_leaf = 0
for i in range(1,20):
    # Kfold = 5
    kfold = StratifiedKFold(n_splits=5, random_state=RANDOM_STATE, shuffle=True)
    score = []
    for idx_train, idx_test in kfold.split(X_Top10_Features, y):
        X_train = X_Top10_Features[idx_train]
        X_test = X_Top10_Features[idx_test]
        y_train = y[idx_train]
        y_test = y[idx_test]
        model = DecisionTreeClassifier(max_depth=best_max_depth,min_samples_leaf=i, random_state = RANDOM_STATE)
        model.fit(X_train, y_train)
        score.append(model.score(X_test, y_test))
    mean = np.mean(score)
    res.append(mean)
    if mean > bestScore:
        best_min_samples_leaf = i
        bestScore = mean

plt.plot(range(1,20), res, color = "yellow", label = "min_samples_leaf")
plt.legend()
plt.show()

# tuning min_samples_split
res = []
bestScore = 0
best_min_samples_split = 0
for i in range(2,20):
    # Kfold = 5
    kfold = StratifiedKFold(n_splits=5, random_state=RANDOM_STATE, shuffle=True)
    score = []
    for idx_train, idx_test in kfold.split(X_Top10_Features, y):
        X_train = X_Top10_Features[idx_train]
        X_test = X_Top10_Features[idx_test]
        y_train = y[idx_train]
        y_test = y[idx_test]
        model = DecisionTreeClassifier(max_depth=best_max_depth,min_samples_leaf=best_min_samples_leaf, min_samples_split=i, random_state = RANDOM_STATE)
        model.fit(X_train, y_train)
        score.append(model.score(X_test, y_test))
    mean = np.mean(score)
    res.append(mean)
    if mean > bestScore:
        best_min_samples_split = i
        bestScore = mean

plt.plot(range(2,20), res, color = "yellow", label = "min_samples_split")
plt.legend()
plt.show()

print(f"Best depth = {best_max_depth}, best min_samples_leaf = {best_min_samples_leaf}, best min_sample_split = {best_min_samples_split}\n\n")
score_RF = []
score_PDT = []
score_DT = []
score_DS = []

for i in range(100):
    # step 1: k-fold
    kfold = StratifiedKFold(n_splits=5, random_state=RANDOM_STATE, shuffle=True)
    score1 = []
    score2 = []
    score3 = []
    score4 = []
    for idx_train, idx_test in kfold.split(X_Top10_Features, y):
        X_train = X_Top10_Features[idx_train]
        X_test = X_Top10_Features[idx_test]
        y_train = y[idx_train]
        y_test = y[idx_test]
        # step 2: Random Forest
        score1.append(RandomForestClassifier(n_estimators=20, random_state = RANDOM_STATE, n_jobs=-1).fit(X_train, y_train).score(X_test, y_test))

        # step 3: Pruned Decision Tree, pruning is on validation set

        score2.append(DecisionTreeClassifier(max_depth = best_max_depth, min_samples_leaf=best_min_samples_leaf, min_samples_split=best_min_samples_split).fit(X_train, y_train).score(X_test, y_test))

        # step 4: Unpruned Decision Tree
        score3.append(DecisionTreeClassifier(random_state=RANDOM_STATE).fit(X_train, y_train).score(X_test, y_test))

        # step 5: Dicision stamp
        score4.append(DecisionTreeClassifier(max_depth=1, random_state=RANDOM_STATE).fit(X_train, y_train).score(X_test, y_test))
    score_RF.append(np.mean(score1))
    score_PDT.append(np.mean(score2))
    score_DT.append(np.mean(score3))
    score_DS.append(np.mean(score4))
# step 6: Compare the models
df = pd.DataFrame()
df["Random Forest"] = score_RF
df["Pruned Decision Tree"] = score_PDT
df["Unpruned Decision Tree"] = score_DT
df["Decision Stamp"] = score_DS
print(df)
result = autorank(df,alpha=0.05, verbose=False)
print(result)
create_report(result)
plot_stats(result)
```
